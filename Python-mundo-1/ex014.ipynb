```python
print("=== Conversor de temperatura ===")
print("Escreva um programa que converta uma temperatura digitando em graus Celsius e converta para graus Fahrenheit.\n\n\n")

temperatura = float(input("Digite a temperatura em graus Celsius: "))

fahrenheit = (temperatura * 9 / 5) + 32

print("{}째 Celsius correspondem a {}째 Fahrenheit".format(temperatura, fahrenheit))
```
