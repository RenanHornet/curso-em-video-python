```python
print("=-=-= Procurando uma string dentro da outra =-=-=")
print("Crie um programa que leia o nome de uma pessoa e diga se ela tem “SILVA” no nome.\n\n\n")

nome = str(input("Digite seu nome completo: ")).strip().upper()#Elimina os espaços e deixa tudo em maiúsculo
print("Seu nome tem Silva? {}".format("SILVA" in nome.upper()))#Retorna true se tiver SILVA
```
